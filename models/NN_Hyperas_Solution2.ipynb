```python
from __future__ import print_function
import numpy as np

from hyperopt import Trials, STATUS_OK, tpe
from keras.layers.core import Dense, Dropout, Activation
from keras.models import Sequential
from keras.utils import np_utils

from hyperas import optim
from hyperas.distributions import choice, uniform
```

```python
import pandas as pd
import glob
import pickle
import os
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers import Dense, Dropout, Activation
from tensorflow.python.keras.wrappers.scikit_learn import KerasRegressor
from tensorflow.keras.callbacks import EarlyStopping
from pickle import load
```

```python
def data():
    path = 'https://data-bootcamp-adriana.s3.us-east-2.amazonaws.com/' 
    li = []
    brands = ["Audi","BMW","Ford","Hyundi","Mercedes Benz","Skoda","Toyota","Volkswagen"]
    csv_names = ["audi", "bmw", "ford", "hyundi", "merc", "skoda","toyota", "vw"]
    for i in range(0,8):
        filename = path + csv_names[i] + ".csv"
        print(filename)
        frame = pd.read_csv(filename, index_col=None, header=0)
        frame["make"] = brands[i]
        li.append(frame)
    df = pd.concat(li, axis=0, ignore_index=True)

    df=df.loc[df['year']<2021,:]
    df=df.drop(['tax'],axis=1)

    X = df.drop(['price'],axis=1)
    y = df["price"].values.reshape(-1, 1)
    from sklearn.model_selection import train_test_split
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
    from pickle import load
    scaler = load(open('axis_transformed_v3.pkl', 'rb'))
    X_test_scaled= scaler.transform(X_test).toarray()
    X_train_scaled=scaler.transform(X_train).toarray()
    from sklearn.preprocessing import StandardScaler
    y_scaler = StandardScaler().fit(y_train)
    y_test_scaled = y_scaler.transform(y_test)
    y_train_scaled = y_scaler.transform(y_train)
    return(X_train_scaled, y_train_scaled, X_test_scaled, y_test_scaled )
```

```python
def create_model(X_train_scaled,y_train_scaled,X_test_scaled,y_test_scaled):
    from tensorflow.keras.callbacks import EarlyStopping
    NNmodel = Sequential()
    NNmodel.add(Dense(100, input_shape=(194,), kernel_initializer='normal', activation='relu'))
    NNmodel.add(Dropout({{uniform(0, 1)}}))
    NNmodel.add(Dense({{choice([50, 100, 200])}}))
    NNmodel.add(Activation({{choice(['relu', 'sigmoid'])}}))
    NNmodel.add(Dropout({{uniform(0, 1)}}))
    NNmodel.add(Dense(1, activation='linear'))
    NNmodel.compile(loss='mse', optimizer={{choice(['rmsprop', 'adam', 'sgd'])}}, metrics=['mse','mae'])
#     callback=EarlyStopping(monitor='loss', patience=3)
    result=NNmodel.fit(X_train_scaled, y_train_scaled, epochs={{choice([50, 100, 200])}}, batch_size={{choice([50, 100])}},  verbose=2, validation_split=0.2)#callbacks=[callback])
    validation_mse = np.amin(result.history['val_mse']) 
    print('Best validation mse of epoch:', validation_mse)

    return {'loss': validation_mse, 'status': STATUS_OK, 'model': NNmodel}
```

```python
best_run, best_model = optim.minimize(model=create_model,
                                      data=data,
                                      algo=tpe.suggest,
                                      max_evals=5,
                                      trials=Trials(),
                                      notebook_name='Cars_Evaluation_Hyperas')
X_train_scaled, y_train_scaled, X_test_scaled, y_test_scaled = data()
print("Evalutation of best performing model:")
print(best_model.evaluate(X_test_scaled, y_test_scaled))
print("Best performing model chosen hyper-parameters:")
print(best_run)
```
